```python
import numpy as np
import matplotlib.pyplot as plt
from mymath import power, factorial
```

```python
def maclaurin_cos(x, terms=10):
    result = 0
    for n in range(terms):
        term = power(-1, n) * power(x, 2 * n) / factorial(2 * n)
        result += term
    return result
```

```python
print(maclaurin_cos(5, 10))
```

```python
x_values = np.linspace(-2 * np.pi, 2 * np.pi, 400)
```

```python
y_values = [maclaurin_cos(x, terms=10) for x in x_values]
y_exact = np.cos(x_values)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='Maclaurin Series Approximation', linestyle='--')
plt.plot(x_values, y_exact, label='Exact Cosine', color='r')
plt.title('Maclaurin Series Approximation of Cos(x)')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.legend()
plt.grid(True)
plt.show()
```
