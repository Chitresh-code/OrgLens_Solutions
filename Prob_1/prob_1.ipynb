```python
import csv
import matplotlib.pyplot as plt
```

### Reading CSV File

```python
data = []
with open(r'D:\OrgLens\Prob_1\marks.csv', mode='r') as file:
    text = csv.DictReader(file)
    for row in text:
        data.append({'Name': row['Name'], 'Marks': int(row['Marks'])})
```

```python
print(data)
```

```python
marks = [mark['Marks'] for mark in data]
print(marks)
```

### Calculating Mean, Variance and SD

```python
mean = sum(marks) / len(marks)
variance = sum((x - mean) ** 2 for x in marks) / len(marks)
std_dev = variance ** 0.5
```

```python
print(f"Mean: {mean:.2f}\nVariance: {variance:.2f}\nSD: {std_dev:.2f}")
```

### Writing in CSV file

```python
with open(r'D:\OrgLens\Prob_1\marks.csv', 'a', newline='') as f:
    write = csv.writer(f)
    write.writerow(['...', '...'])
    write.writerow(['Mean', f"{mean:.2f}"])
    write.writerow(['Variance', f"{variance:.2f}"])
    write.writerow(['Standard Deviation', f"{std_dev:.2f}"])
```

### Plotting

```python
# Plotting the marks
plt.plot(marks, marker='o', linestyle='-', color='b', label='Marks')

# Marking the Mean
plt.axhline(y=mean, color='g', linestyle='--', label=f'Mean: {mean:.2f}')

# Marking the Standard Deviation
plt.axhline(y=mean + std_dev, color='r', linestyle='--', label=f'Mean + SD: {mean + std_dev:.2f}')
plt.axhline(y=mean - std_dev, color='r', linestyle='--', label=f'Mean - SD: {mean - std_dev:.2f}')

# Add labels and legend
plt.xlabel('Student Index')
plt.ylabel('Marks')
plt.title('Marks Distribution with Mean and Standard Deviation')
plt.legend()
```
